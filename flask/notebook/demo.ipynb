```python
# Tổng hợp theo cả luồng và loại tờ khai
tong_hop_ket_hop = df.groupby(['luong', 'loai_to_khai']).agg({
    'tri_gia': ['sum', 'mean', 'count'],
    'thue_suat': 'mean'
}).round(2)
tong_hop_ket_hop.columns = ['Tổng trị giá', 'Trị giá TB', 'Số lượng', 'Thuế suất TB']
print("Tổng hợp theo Luồng và Loại tờ khai:")
tong_hop_ket_hop
```
