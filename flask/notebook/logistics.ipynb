```python
import pandas as pd
```

```python
# Đọc dữ liệu từ file CSV
df = pd.read_csv('logistics.csv')

# Hiển thị thông tin cơ bản về dữ liệu
print("Kích thước dữ liệu:", df.shape)
print("\n" + "="*50)
print("Các cột trong dataset:")
print(df.columns.tolist())
print("\n" + "="*50)
print("Thông tin về kiểu dữ liệu:")
print(df.info())
print("\n" + "="*50)
print("5 dòng đầu tiên:")
df.head()
```

## Phân tích thống kê mô tả

```python
# Thống kê mô tả cho các cột số
print("THỐNG KÊ MÔ TẢ:")
print(df.describe())
print("\n" + "="*50)

# Kiểm tra giá trị null
print("\nGIÁ TRỊ NULL:")
print(df.isnull().sum())
print("\n" + "="*50)

# Thống kê giá trị duy nhất
print("\nSỐ LƯỢNG GIÁ TRỊ DUY NHẤT:")
print(df.nunique())
```

## Phân tích theo danh mục

```python
# Phân tích theo quốc gia
print("PHÂN PHỐI THEO QUỐC GIA:")
print(df['Country'].value_counts())
print("\n" + "="*50)

# Phân tích theo cảng
print("\nPHÂN PHỐI THEO CẢNG:")
print(df['Port'].value_counts())
print("\n" + "="*50)

# Phân tích theo sản phẩm
print("\nPHÂN PHỐI THEO SẢN PHẨM:")
print(df['Product'].value_counts())
print("\n" + "="*50)

# Phân tích theo phương thức vận chuyển
print("\nPHÂN PHỐI THEO PHƯƠNG THỨC VẬN CHUYỂN:")
print(df['TransportMode'].value_counts())
print("\n" + "="*50)

# Phân tích theo trạng thái
print("\nPHÂN PHỐI THEO TRẠNG THÁI:")
print(df['Status'].value_counts())
```

## Phân tích nâng cao

```python
# Tổng giá trị theo quốc gia
print("TỔNG GIÁ TRỊ (USD) THEO QUỐC GIA:")
country_value = df.groupby('Country')['Value(USD)'].sum().sort_values(ascending=False)
print(country_value)
print(f"\nTổng giá trị: ${country_value.sum():,.2f}")
print("\n" + "="*50)

# Tổng khối lượng theo sản phẩm
print("\nTỔNG KHỐI LƯỢNG (KG) THEO SẢN PHẨM:")
product_qty = df.groupby('Product')['Quantity(kg)'].sum().sort_values(ascending=False)
print(product_qty)
print(f"\nTổng khối lượng: {product_qty.sum():,.2f} kg")
print("\n" + "="*50)

# Thời gian thông quan trung bình theo phương thức vận chuyển
print("\nTHỜI GIAN THÔNG QUAN TRUNG BÌNH (NGÀY) THEO PHƯƠNG THỨC VẬN CHUYỂN:")
clearance_time = df.groupby('TransportMode')['ClearanceTime(days)'].mean().sort_values(ascending=False)
print(clearance_time)
print("\n" + "="*50)

# Giá trị trung bình theo trạng thái
print("\nGIÁ TRỊ TRUNG BÌNH (USD) THEO TRẠNG THÁI:")
status_value = df.groupby('Status')['Value(USD)'].mean().sort_values(ascending=False)
print(status_value)
```

## Phân tích đa chiều

```python
# Bảng tổng hợp theo quốc gia và sản phẩm
print("TỔNG GIÁ TRỊ THEO QUỐC GIA VÀ SẢN PHẨM:")
pivot_country_product = pd.pivot_table(
    df, 
    values='Value(USD)', 
    index='Country', 
    columns='Product', 
    aggfunc='sum', 
    fill_value=0
)
print(pivot_country_product)
print("\n" + "="*50)

# Top 10 lô hàng có giá trị cao nhất
print("\nTOP 10 LÔ HÀNG CÓ GIÁ TRỊ CAO NHẤT:")
top_shipments = df.nlargest(10, 'Value(USD)')[['ShipmentID', 'Country', 'Product', 'Value(USD)', 'Status']]
print(top_shipments)
```

## Trực quan hóa dữ liệu với Plotly

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Đếm số lượng theo trạng thái
status_counts = df['Status'].value_counts().reset_index()
status_counts.columns = ['Status', 'Count']

# Vẽ biểu đồ cột so sánh trạng thái
fig = px.bar(
    status_counts,
    x='Status',
    y='Count',
    title='Biểu đồ so sánh số lượng lô hàng theo trạng thái',
    labels={'Status': 'Trạng thái', 'Count': 'Số lượng lô hàng'},
    color='Status',
    color_discrete_map={
        'Delivered': '#2ecc71',
        'In Transit': '#3498db',
        'Delayed': '#e74c3c',
        'Hold': '#f39c12'
    },
    text='Count'
)

# Tùy chỉnh layout
fig.update_traces(textposition='outside')
fig.update_layout(
    showlegend=True,
    height=500,
    font=dict(size=12),
    xaxis_title='Trạng thái',
    yaxis_title='Số lượng lô hàng'
)

fig.show()
```

## Bảng thống kê theo quốc gia

```python
# Tạo bảng thống kê theo quốc gia
country_stats = df.groupby('Country').agg({
    'ShipmentID': 'count',
    'Value(USD)': ['sum', 'mean', 'min', 'max'],
    'Quantity(kg)': 'sum'
}).round(2)

# Đổi tên cột
country_stats.columns = ['Số lượng chuyến hàng', 'Tổng giá trị (USD)', 'Giá trị TB (USD)', 'Giá trị min (USD)', 'Giá trị max (USD)', 'Tổng khối lượng (kg)']

# Sắp xếp theo tổng giá trị giảm dần
country_stats = country_stats.sort_values('Tổng giá trị (USD)', ascending=False)

# Reset index để Country thành cột
country_stats = country_stats.reset_index()

print("BẢNG THỐNG KÊ CHUYẾN HÀNG THEO QUỐC GIA")
print("="*100)
country_stats
```

## Bảng thống kê theo cảng và phương thức vận chuyển

```python
# Tạo bảng thống kê theo cảng và phương thức vận chuyển
port_transport_stats = df.groupby(['Port', 'TransportMode']).agg({
    'ShipmentID': 'count',
    'Value(USD)': ['sum', 'mean'],
    'Quantity(kg)': 'sum',
    'ClearanceTime(days)': 'mean'
}).round(2)

# Đổi tên cột
port_transport_stats.columns = [
    'Số lượng chuyến hàng',
    'Tổng giá trị (USD)',
    'Giá trị TB (USD)',
    'Tổng khối lượng (kg)',
    'Thời gian thông quan TB (ngày)'
]

# Sắp xếp theo tổng giá trị giảm dần
port_transport_stats = port_transport_stats.sort_values('Tổng giá trị (USD)', ascending=False)

# Reset index
port_transport_stats = port_transport_stats.reset_index()

print("BẢNG THỐNG KÊ CHUYẾN HÀNG THEO CẢNG VÀ PHƯƠNG THỨC VẬN CHUYỂN")
print("="*120)
port_transport_stats
```

```python
# Thống kê tổng hợp theo cảng
port_stats = df.groupby('Port').agg({
    'ShipmentID': 'count',
    'Value(USD)': ['sum', 'mean', 'min', 'max'],
    'Quantity(kg)': 'sum',
    'ClearanceTime(days)': 'mean'
}).round(2)

# Đổi tên cột
port_stats.columns = [
    'Số lượng chuyến hàng',
    'Tổng giá trị (USD)',
    'Giá trị TB (USD)',
    'Giá trị min (USD)',
    'Giá trị max (USD)',
    'Tổng khối lượng (kg)',
    'Thời gian thông quan TB (ngày)'
]

# Sắp xếp theo tổng giá trị giảm dần
port_stats = port_stats.sort_values('Tổng giá trị (USD)', ascending=False)

# Reset index
port_stats = port_stats.reset_index()

print("\nBẢNG THỐNG KÊ TỔNG HỢP THEO CẢNG")
print("="*120)
port_stats
```
