```python
import pandas as pd

import pymssql

MSSQL_CONFIG = {
    'server': 'server.softech.cloud',
    'database': 'OnlineShop',
    'user': 'tester',
    'password': '38yenbai'
}

def get_db_connection():
    """Tạo kết nối đến MSSQL"""
    conn = pymssql.connect(
        server=MSSQL_CONFIG['server'],
        user=MSSQL_CONFIG['user'],
        password=MSSQL_CONFIG['password'],
        database=MSSQL_CONFIG['database']
    )
    return conn



# Read data from a CSV file into a DataFrame
data = pd.read_csv('danh-sach-tau.csv')

# Display the first few rows of the DataFrame
print(data.head())

# Tôi có 1 connection đến MSQL Server
conn = get_db_connection()

# Tôi muốn chèn dữ liệu từ DataFrame vào bảng có tên là 'danh_sach_tau' trong database
# Sử dụng phương thức to_sql của pandas để chèn dữ liệu
# Dùng câu lệnh SQL INSERT để chèn dữ liệu từ DataFrame vào bảng
cursor = conn.cursor()
for index, row in data.iterrows():
    cursor.execute("""
        INSERT INTO danh_sach_tau (name, code, port_to, port_from) 
        VALUES (%s, %s, %s, %s)
    """, (row['name'], row['code'], row['port_to'], row['port_from']))
conn.commit()
cursor.close()# Kết thúc kết nối
conn.close()
```
